### Очень сложная программа с функциями

```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def find_owner(my_doc,documents):
    """
    Нужно ввести номер документа, по нем найдется владелец
    """
    flag = 0 
    for doc in documents: 
        if doc['number'] == my_doc:
            print(f'Владелец документа: {doc["name"]}')
            flag = 1 
            return 1 # 
            break # если есть return, то мне уже не нужен break, правильно?
    if flag == 0:
        return 0
        print('Документ не найден!') # итого, функция будет и выводить текст, и возвращать 1 или 0, по которым я могу проверяться в будущем? То есть проверять, есть ли такой документ вообще.

def find_type_doc(my_doc,documents):
    """
    Нужно ввести номер документа, по нем найдется тип документа
    """
    flag = 0 
    for doc in documents: 
        if doc['number'] == my_doc:
            print(f'Тип документа: {doc["type"]}')
            flag = 1 
            break
    if flag == 0:
        print('Документ не найден!')
        
def find_doc(my_doc,directories): 
    """
    Нужно ввести номер документа, по нем найдется полка
    """
    flag = 0
    for polka in directories:
        if my_doc in directories[polka]:
            flag = 1
            print (f'На полке №{polka}')
            break
    if flag == 0:
        print('Документ не найден!')

def show_all(documents,directories):
    """
    показывает все данные в реестре
    """
    list_doc = []
    for doc in documents: 
         list_doc.append(doc['number'])

    for el in list_doc:
        print(f'Номер документа: {el}') 
        find_type_doc(el,documents) 
        find_owner(el,documents)
        find_doc(el,directories)
        print()
        
def is_empty_polka(num,directories):
    """
    проверяет пуста ли полка, нужно внести номер полки
    """
    if directories.get(num) == []:
        return 1 
    else:
        return 0

def polka_operation(id_op, num_polka, directories):
    """
    удаляет или добавляет полку: сначала внести операцию, потом номер полки
    """
    len_d = len(directories)
    if id_op == 'as': 
        directories.setdefault(num_polka, [])
        if len(directories) > len_d:
            print(f'Полка добавлена. Текущий перечень полок: {", ".join(list(directories.keys()))}')
        else:
            print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(list(directories.keys()))}')
    if id_op == 'ds':
        if num_polka not in directories:
            print(f'Полки не существует. Текущий перечень полок: {", ".join(list(directories.keys()))}')
        elif is_empty_polka(num_polka,directories) == 1:
            del(directories[num_polka])
            if len(directories) < len_d:
                print(f'Полка удалена. Текущий перечень полок: {", ".join(list(directories.keys()))}')
        else:
            print(f'Перед удалением полку нужно очистить. Текущий перечень полок: {", ".join(list(directories.keys()))}') 

def add_doc(documents,directories):
    """
    Добавляет в базу и на полку новый документ
    """
    doc_id = input('Введте номер документа ')
    doc_type = input('Введте тип документа ')
    doc_owner = input('Введте владельца документа ')
    num_polka = input('Введте полку для хранения ')
    print()

    if num_polka not in directories:
        print('Такой полки не существует. Добавьте полку командой "as".')
        print('Текущий список документов:')
        show_all(documents,directories)
    else:
        documents.append({'type': '', 'number': '', 'name': '' })
        documents[len(documents)-1]['number'] = doc_id
        documents[len(documents)-1]['type'] = doc_type
        documents[len(documents)-1]['name'] = doc_owner
        directories[num_polka].append(doc_id)
        print('Документ добавлен. Текущий список документов:')
        show_all(documents,directories)
```

```python
#Главная функция
def main():
    """
    удаляет или добавляет полку: сначала внести операцию, потом номер полки
    """
    while True:
        command = input('Введте команду ')
        if command ==  'p':
            my_doc = input('Введте номер документа ')
            find_owner(my_doc,documents)
        elif command == 's':
            my_doc = input('Введте номер документа ')
            find_doc(my_doc,documents)
        elif command == 'l':
            show_all(documents,directories)
        elif command ==  'as':
            num_polka = input('Введте номер полки: ')
            polka_operation(command, num_polka, directories)
        elif command ==  'ds':
            num_polka = input('Введте номер полки: ')
            polka_operation(command, num_polka, directories)
        elif command ==  'ad':
            add_doc(documents,directories)
        elif command ==  'q':
            print ('Вы вышли из программы')
            break
        else:
            print ('Такой команды нет')

# command = ''
# while command != 'q':
#     command = input('Введте команду ')
#     main(command)
main()


    
```
